# K-Means Clustering implementation


```
from model import Model
from sklearn.datasets import load_iris
data = load_iris()
```

```
model = Model(3)
model.fit(data['data'])
for point in model.points:
    print('POINT COORDS', point.coordinates)
    print('CENTROID COORDS', point.assigned_centroid.coordinates)
    print('--------------------')

print('Model prediction for [2,2,2,2]', model.predict([2,2,2,2]))
```

# Sklearn implementation

```
from sklearn.cluster import KMeans
import numpy as np

X = data['data']
kmeans = KMeans(n_clusters=3).fit(X)


print('predicted cluster:', kmeans.predict([[2,2,2,2]]))

kmeans.cluster_centers_[kmeans.predict([[2,2,2,2]])]
```

As you can see the sklearn implementation predicts the same centroid/cluster as the implementation we made.
